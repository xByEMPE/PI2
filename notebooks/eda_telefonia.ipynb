```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
telefonia_movil = pd.ExcelFile('C:/PI2/datasets/Telefonia_movil.xlsx')
```

# Cargamos todas las hojas en un diccionario de DataFrames


```python
sheets_data = {sheet: telefonia_movil.parse(sheet) for sheet in telefonia_movil.sheet_names}
               
```

### Generamos la grafica para representar tendencias por periodo

```python
def plot_trends(data, x_col, y_cols, title, ylabel):
    plt.figure(figsize=(10, 6))
    for col in y_cols:
        plt.plot(data[x_col], data[col], marker='o', label=col)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.title(title)
    plt.xlabel('Periodo')
    plt.ylabel(ylabel)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

### Preparamos y ordenamos los datos para despues graficarlos

```python
for sheet_name in ['SMS', 'Llamadas salientes', 'Minutos salientes', 'Ingresos', 'Penetracion', 'Accesos']:
    sheets_data[sheet_name] = sheets_data[sheet_name].sort_values(by=['AÃ±o', 'Trimestre'])
```

## Grafica tendencia de SMS salientes

```python
plot_trends(
    sheets_data['SMS'],
    x_col='Periodo',
    y_cols=['NÃºmero de SMS salientes'],
    title='Tendencia de SMS salientes por perÃ­odo',
    ylabel='NÃºmero de SMS salientes'
)
```

- Se observa una disminucion constante en el numero de SMS enviados, probablemente sea al aumento del uso de servicios de mensajeria basados en internet

## Tendencia de llamadas salientes

```python
plot_trends(
    sheets_data['Llamadas salientes'],
    x_col='Periodo',
    y_cols=[
        'Llamadas pospago salientes (miles)',
        'Llamadas prepago salientes (miles)',
        'Total de llamadas salientes (miles)'
    ],
    title='Tendencia de llamadas salientes por perÃ­odo',
    ylabel='NÃºmero de llamadas (miles)'
)
```

- Los datos indican varaiciones entre llamadas pospago y prepago, con un ligero aumento en el total hasta alcanzar un punto de estabilizacion 

## Tendencia de minutos salientes

```python
plot_trends(
    sheets_data['Minutos salientes'],
    x_col='Periodo',
    y_cols=[
        'Minutos pospago salientes (miles)',
        'Minutos prepago salientes (miles)',
        'Total de minutos salientes (miles)'
    ],
    title='Tendencia de minutos salientes por perÃ­odo',
    ylabel='NÃºmero de minutos (miles)'
)
```

- Se observa una tendencia similar al de las llamadas, donde los minutos pospago tienen un peso significativo en el total 

## Tendencia de ingresos

```python
plot_trends(
    sheets_data['Ingresos'],
    x_col='Periodo',
    y_cols=['Ingresos (miles de $)'],
    title='Tendencia de ingresos por perÃ­odo',
    ylabel='Ingresos (miles de $)'
)
```

- La grafica es clara y muestra una tendencia de crecimiento, reflejando una expansion del mercado o aumentos tarifarios

## Tendencia de penetracion 

```python
plot_trends(
    sheets_data['Penetracion'],
    x_col='Periodo',
    y_cols=['Accesos por cada 100 hab'],
    title='Tendencia de penetraciÃ³n por perÃ­odo',
    ylabel='Accesos por cada 100 habitantes'
)
```

- Los accesos por cada 100 habitantes tiene una tendencia ascendente hasta el 2013, pero luego disminuyen, posiblemente debido a cambios en las politicas del mercado o la saturacion.

## Tendencias de accesos

```python
plot_trends(
    sheets_data['Accesos'],
    x_col='Periodo',
    y_cols=[
        'Total de accesos pospago',
        'Total de accesos prepago',
        'Total de accesos operativos'
    ],
    title='Tendencia de accesos por perÃ­odo',
    ylabel='NÃºmero de accesos'
)
```

- Los accesos prepago dominan, aunque muestran una reduccion en los ultimos periodos, lo que podria estar vinculado a una transicion hacia planes pospag.
