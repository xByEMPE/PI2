```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:/PI2/datasets/Television.xlsx'  # Reemplaza con la ruta a tu archivo
excel_data = pd.ExcelFile(file_path)
```

```python
sheets_data = {sheet: excel_data.parse(sheet) for sheet in excel_data.sheet_names}
```

```python
# Análisis de la hoja "Accesos_totales_TV"
total_accesses = sheets_data['Accesos_totales_TV']
total_accesses_sorted = total_accesses.sort_values(by=['Año', 'Trimestre'])
```

```python
# Graficar tendencia de accesos totales a TV
plt.figure(figsize=(10, 6))
plt.plot(
    total_accesses_sorted['Año'].astype(str) + 'Q' + total_accesses_sorted['Trimestre'].astype(str),
    total_accesses_sorted['Accesos TV por suscripción'],
    label='TV por Suscripción'
)
plt.plot(
    total_accesses_sorted['Año'].astype(str) + 'Q' + total_accesses_sorted['Trimestre'].astype(str),
    total_accesses_sorted['Accesos TV satelital'],
    label='TV Satelital'
)
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.title('Tendencia de Accesos Totales a TV por Año y Trimestre')
plt.xlabel('Año y Trimestre')
plt.ylabel('Accesos')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Análisis de la hoja "Ingresos_TV"
revenues = sheets_data['Ingresos_TV']
revenues_sorted = revenues.sort_values(by=['Año', 'Trimestre'])
```

```python
plt.figure(figsize=(10, 6))
plt.bar(
    revenues_sorted['Año'].astype(str) + 'Q' + revenues_sorted['Trimestre'].astype(str),
    revenues_sorted['Ingresos TV por suscripción  (miles de $)'],
    label='Ingresos TV por Suscripción'
)
plt.bar(
    revenues_sorted['Año'].astype(str) + 'Q' + revenues_sorted['Trimestre'].astype(str),
    revenues_sorted['Ingresos TV satelital (miles de $)'],
    bottom=revenues_sorted['Ingresos TV por suscripción  (miles de $)'],
    label='Ingresos TV Satelital'
)
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.title('Ingresos Totales por Tipo de TV y Periodo')
plt.xlabel('Periodo (Año y Trimestre)')
plt.ylabel('Ingresos (miles de $)')
plt.legend()
plt.tight_layout()
plt.show()
```
