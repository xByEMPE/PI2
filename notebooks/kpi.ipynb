# Calculo del KPI 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
internet_path = 'C:/PI2/datasets/Internet.xlsx'
internet_data = pd.ExcelFile(internet_path)
```

### Utilizaremos la hoja de 'Penetracion-hogares' para este KPI

```python
penetracion_hogares = internet_data.parse('Penetracion-hogares')
```

### Calculamos el KPI con un incremento del 2% en accesos

```python
penetracion_hogares['Nuevo acceso'] = penetracion_hogares['Accesos por cada 100 hogares'] * 1.02
penetracion_hogares['KPI (%)'] = ((penetracion_hogares['Nuevo acceso'] - 
                                   penetracion_hogares['Accesos por cada 100 hogares']) / 
                                  penetracion_hogares['Accesos por cada 100 hogares']) * 100
```

### Se procede a graficar el KPI por provincia 

```python
plt.figure(figsize=(14, 8))
plt.bar(penetracion_hogares['Provincia'], penetracion_hogares['Accesos por cada 100 hogares'], 
        color='blue', label='Acceso Actual')
plt.bar(penetracion_hogares['Provincia'], penetracion_hogares['Nuevo acceso'], 
        color='green', alpha=0.6, label='Nuevo Acceso (Meta)')
plt.title('Comparación de Accesos Actuales vs Meta por Provincia', fontsize=16)
plt.xlabel('Provincia', fontsize=12)
plt.ylabel('Accesos por cada 100 hogares', fontsize=12)
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
