```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "C:/PI2/datasets/Telefonia_movil.xlsx"
```

```python
ingresos_df = pd.read_excel(file_path, sheet_name="Ingresos")
ingresos_df['Periodo'] = ingresos_df['Periodo'].str.strip()
```

```python
ingresos_df['Crecimiento (%)'] = ingresos_df['Ingresos (miles de $)'].pct_change() * 100
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ingresos_df['Periodo'], ingresos_df['Crecimiento (%)'], marker='o', label="Crecimiento (%)")
plt.axhline(0, color='red', linestyle='--', label='Crecimiento = 0%')
plt.title("Evolucin trimestral del crecimiento de ingresos (%)")
plt.xlabel("Periodo")
plt.ylabel("Crecimiento (%)")
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print("Datos calculados:")
print(ingresos_df[['Periodo', 'Ingresos (miles de $)', 'Crecimiento (%)']])
```

# Analisis del KPI 
Porqu en algunos tirmestres vemos valores negativos?
Los valores negativos en el KPI del crecimiento trimestral de ingresos se deben a una disminucin de los ingresos en un trimestre en comparacin con el trimestre anterior

Este comportamiento se calcula mediante la frmula del KPI:

Crecimiento(%) = (Ingresos actuales - Ingresos anteriores / Ingresos anteriores) * 100 
Cuando los ingresos actuales son menores que los ingresos del trimestre anterior, el numerador de la frmula ser negativo, resultando en un porcentaje de crecimiento negativo. Esto indica una contraccin o una prdida en los ingresos durante ese perodo.

### **Causas posibles**:
* Estacionalidad: Algunos trimestres pueden tener menores ingresos debido a cambios estacionales en la demanda
* Factores econmicos: Cambios en la economa pueden haber reducido el consumo de servicios mviles.
* Competencia: Prdida de mercado frente a competidores.
* Errores en los datos: Es posible que los valores negativos se deban a inconsistencias o errores en los datos.
