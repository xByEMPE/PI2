# Calculo del KPI 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
internet_path = 'C:/PI2/datasets/Internet.xlsx'
internet_data = pd.ExcelFile(internet_path)
```

### Utilizaremos la hoja de 'Penetracion-hogares' para este KPI

```python
penetracion_hogares_df = internet_data.parse('Penetracion-hogares')
```

### Calculamos el KPI con un incremento del 2% en accesos

```python
penetracion_hogares_df['Projected Access (Next Quarter)'] = penetracion_hogares_df['Accesos por cada 100 hogares'] * 1.02
penetracion_hogares_df['KPI Actual'] = (
    (penetracion_hogares_df['Projected Access (Next Quarter)'] - 
     penetracion_hogares_df['Accesos por cada 100 hogares']) / 
    penetracion_hogares_df['Accesos por cada 100 hogares']
) * 100
```

### Se procede a graficar el KPI por provincia 

```python
plt.figure(figsize=(12, 8))
plt.barh(penetracion_hogares_df['Provincia'], penetracion_hogares_df['Projected Access (Next Quarter)'], label='Acceso proyectado', color='blue')
plt.barh(penetracion_hogares_df['Provincia'], penetracion_hogares_df['Accesos por cada 100 hogares'], label='Acceso actual', color='orange', alpha=0.7)
plt.xlabel('Acceso por cada 100 hogares')
plt.ylabel('Provincia')
plt.title('Proyectado vs Acceso actual por cada 100 hogares por provincia') 
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
```

- Esto muestra la comparativa en un aumento del 2% en el acceso a internet en las provincias para el proximo trimestre.
